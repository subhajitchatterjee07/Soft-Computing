```python
# generate 100 random values for 5 variables
a = np.random.rand(100)
b = np.random.rand(100)
c = np.random.rand(100)
d = np.random.rand(100)
e = np.random.rand(100)

y = a**2 + b**2 + c**2 + d**2 + e**2

data = np.column_stack((a, b, c, d, e, y))
print(data[:5])
```
