#  BACKPROPAGATED FEEDFORWARD NEURAL NETWORK

```python


def activation_fn(x):
    return 1 / (1 + np.exp(-x))

def activation_derivative(x):
    return activation_fn(x) * (1 - activation_fn(x))

alpha = 0.1
w_ih = np.random.rand(5, 7)  # weight matrix from input layer to hidden layer
res1 = np.random.rand(1, 7)  # Initialize res1 as a row vector
w_ho = np.random.rand(7, 1)  # weight matrix from hidden layer to output layer
res2 = np.random.rand(1, 1)  # Initialize res2 as a scalar
b_ih = np.random.rand(1, 7)  # bias for hidden layer
b_ho = np.random.rand(1, 1)  # bias for output layer

for k, rw in enumerate(data[:, :5]):
    rw = np.atleast_2d(rw)
    
    # Beginning of the feed forward phase
    print(res1.shape)
    print(rw.shape)
    print(w_ih.shape)
    
    # Step 1 - from input layer to hidden layer
    res1 = np.dot(rw, w_ih[:, i]) + b_ih
    print(res1)
    act_arr1 = activation_fn(res1)
    
    # Step 2 - from hidden layer to output layer
    res2 = np.dot(act_arr1, w_ho) + b_ho
    print(res2)
    act_arr2 = activation_fn(res2)

    # Backpropagation begins
    print("___________________________________________")
    # Output to hidden layer
    del_k = (data[k, -1] - act_arr2) * activation_derivative(res2)
    del_w_ho = alpha * np.dot(act_arr1.T, del_k)
    del_b_ho = alpha * np.sum(del_k)
    
    # Hidden to input layer
    del_in_j = np.dot(del_k, w_ho.T)
    del_j = del_in_j * activation_derivative(res1)
    del_w_ih = alpha * np.dot(rw.T, del_j)
    del_b_ih = alpha * np.sum(del_j)
    print(del_w_ih.shape)
    print(del_w_ho.shape)

    # Updating the weights
    w_ho += del_w_ho
    w_ih += del_w_ih
    b_ho += del_b_ho
    b_ih += del_b_ih
    print("___________________________________________")

```
