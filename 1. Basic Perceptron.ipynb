## AND gate implementation using a Perceptron

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class perceptron:
    def __init__(self,input_size):
        # Step1: create a set of w's and b's for the neuron
        self.weights=np.random.rand(input_size)
        self.bias=np.random.rand(1)
    #step2 Activate function
    def activate(self,x):
        return 1 if x>0 else 0
    
    def forward(self,inputs):
        # perform forward pass
        weighted_sum = np.dot(inputs,self.weights)+self.bias
        output = self.activate(weighted_sum)
        return output
    
    def train(self,inputs,targets,epochs,learning_rate):
        for epoch in range(epochs):
            for input_data,target in zip(inputs,targets):
                input_data=np.array(input_data,ndmin=1)

                #forward pass
                pred=self.forward(input_data)

                #update weights and bias
                error=target-pred
                self.weights+=learning_rate*error*input_data
                self.bias+=learning_rate*error
            
            if epoch%100==0:
                accuracy=np.mean(targets == np.array([self.forward(x) for x in inputs]))
                print(f"Epoch {epoch}, Accuracy: {accuracy}")
        
```

```python
inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
targets = np.array([0, 0, 0, 1])

# Create a perceptron with 2 input nodes
percep = perceptron(input_size=2)

# Train the perceptron
percep.train(inputs, targets, epochs=1000, learning_rate=0.1)
```

```python
for input_data in inputs:
    prediction = percep.forward(input_data)
    print(f"Input: {input_data}, Prediction: {prediction}")
```

## NAND Gate Implementation with a Perceptron

```python
import numpy as np

class Perceptron:
    def __init__(self, input_size):
        # Initialize weights and bias
        self.weights = np.random.rand(input_size)
        self.bias = np.random.rand(1)

    def step_function(self, x):
        # Step activation function
        return 1 if x > 0 else 0

    def forward(self, inputs):
        # Perform forward pass
        weighted_sum = np.dot(inputs, self.weights) + self.bias
        output = self.step_function(weighted_sum)
        return output

    def train(self, inputs, targets, epochs, learning_rate):
        # Train the perceptron
        for epoch in range(epochs):
            for input_data, target in zip(inputs, targets):
                input_data = np.array(input_data, ndmin=1)
                target = np.array(target, ndmin=1)

                # Forward pass
                prediction = self.forward(input_data)

                # Update weights and bias
                error = target - prediction
                self.weights += learning_rate * error * input_data
                self.bias += learning_rate * error

            if epoch % 100 == 0:
                accuracy = np.mean(targets == np.array([self.forward(x) for x in inputs]))
                print(f"Epoch {epoch}, Accuracy: {accuracy}")

# Example usage for AND gate
inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1], [0.4,0.4]])
targets = np.array([1, 1, 1, 0, 0])

# Create a perceptron with 2 input nodes
perceptron = Perceptron(input_size=2)

# Train the perceptron
perceptron.train(inputs, targets, epochs=1100, learning_rate=0.1)

# Test the trained perceptron
for input_data in inputs:
    prediction = perceptron.forward(input_data)
    print(f"Input: {input_data}, Prediction: {prediction}")
```
